```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
California_Texas_2017 = df[ ((df["State"] == "California") | (df["State"] == "Texas")) & (df["Order Date"].dt.year == 2017) ]
California_Texas_2017  
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
data_2017 = df[df["Order Date"].dt.year == 2017]["Sales"].agg(["sum","mean","std"])

print("total sales = ",data_2017[0],
      "average sales = ",data_2017[1],
      "standard deviation of sales = ",data_2017[2])
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df["Order Date"].dt.year == 2018].groupby("Segment")["Profit"].sum().reset_index()
df_2018.sort_values("Profit",ascending=False).head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df_2019 = df[(df["Order Date"] >= "2019-04-15") & (df["Order Date"] < "2019-12-31")]\
    .groupby("State")["Sales"].sum().reset_index().sort_values("Sales").head(5)
df_2019
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
sum_wc = df[((df["Region"] == "West") | (df["Region"] == "Central")) & (df["Order Date"].dt.year == 2019)]["Sales"].sum()
sum_total = df[df["Order Date"].dt.year == 2019]["Sales"].sum()

print("the proportion of total sales in West + Central in 2019 = ",int((sum_wc / sum_total)*100), " %" )
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020


prod_order = df[(df["Order Date"].dt.year == 2019) | (df["Order Date"].dt.year == 2020)].groupby("Product Name")["Order ID"]\
    .count().reset_index().sort_values("Order ID",ascending=False).head(10).reset_index().drop("index",axis=1)

prod_sales = df[(df["Order Date"].dt.year == 2019) | (df["Order Date"].dt.year == 2020)].groupby("Product Name")["Sales"]\
    .sum().reset_index().sort_values("Sales",ascending=False).head(10).reset_index().drop("index",axis=1)

top10 = pd.merge(prod_order,prod_sales,left_index=True,right_index=True)
top10
```
